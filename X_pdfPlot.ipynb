## Exemplarischer Plot: So sieht es aus:

### Hourly Plot

```python
#########################################
myrcparams={'axes.labelsize':'large'}
mystyle = mpf.make_mpf_style(base_mpf_style='yahoo',rc=myrcparams)

weeks = list(set(daten.index.strftime("%W").to_list()))
a_week = daten.loc[(daten.index.strftime("%W") >= weeks[42]) & (daten.index.strftime("%W") <= weeks[44])]
a_week["cumvol"]=a_week["Volume"].cumsum()
a_week["cumvolp"]=(a_week["Close"]*a_week["Volume"]).cumsum()
a_week["VP"]=a_week["cumvolp"]/a_week["cumvol"]

    # Calculate VWAP within each day
a_week['Cumulative_Volume'] = a_week['Volume'].groupby(a_week.index.date).cumsum()
a_week['VWAP'] = (a_week['Close'] * a_week['Volume']).groupby(a_week.index.date).cumsum() / a_week['Cumulative_Volume']

#a_week["VP"]=(a_week["Close"]*a_week["Volume"]).cumsum()/a_week.Volume.cumsum()

ticks = a_week.loc[((a_week.index.hour==8))|((a_week.index.hour==10))].index
ticklocations = [ a_week.index.get_loc(t) for t in ticks]
ticklabels = [ (tick.strftime('%A')[:3]+" "+tick.time().strftime('%H:%M')) for tick in ticks ]

startdays=list(a_week.loc[((a_week.index.hour==8)) | ((a_week.index.hour==10))].index)
a_week["mid"]=(a_week["High"]+a_week["Low"]+a_week["Open"]+a_week["Close"])/4
p1=mpf.make_addplot(a_week["mid"]+a_week["ATR60"])
p2=mpf.make_addplot(a_week["mid"]-a_week["ATR60"])
# p3=mpf.make_addplot(a_week["VP"])
# p5=mpf.make_addplot(a_week["VWAP"])
# p4=mpf.make_addplot(a_week["50dayEMA"])
#p5=mpf.make_addplot(a_week["VP"])
candle_data = mpf.make_addplot(a_week[['Open', 'Close', 'High', 'Low',"Volume"]], type='candle')
plotsadded=[p1,p2,candle_data]

fig , axlist = mpf.plot(a_week, type='candle', style=mystyle, addplot=plotsadded, volume=True,  xrotation = 65,title="TESTETST",figscale=2.0, \
                 figratio=(16, 10), vlines=dict(vlines=startdays, linewidths=0.1,alpha=0.2,colors="green"), returnfig=True)

axlist[-2].xaxis.set_ticks(ticklocations)
axlist[-2].set_xticklabels(ticklabels)
mpf.show()
```
